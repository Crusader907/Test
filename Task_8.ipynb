```python
import requests
import pandas as pd

# URL для запроса книг
books_url = "https://www.anapioficeandfire.com/api/books"

# Получаем данные о книгах
response_books = requests.get(books_url)
if response_books.status_code == 200:
    books_data = response_books.json()  # Преобразуем ответ в JSON
else:
    print(f"Ошибка при запросе книг: {response_books.status_code}")
    books_data = []

# Создаем DataFrame из данных о книгах
books_df = pd.DataFrame(books_data)

# Выводим первые строки DataFrame
print("Информация о книгах:")
display(books_df.head())
```

```python
# URL для запроса домов
houses_url = "https://www.anapioficeandfire.com/api/houses"

# Получаем данные о домах
response_houses = requests.get(houses_url)
if response_houses.status_code == 200:
    houses_data = response_houses.json()  # Преобразуем ответ в JSON
else:
    print(f"Ошибка при запросе домов: {response_houses.status_code}")
    houses_data = []

# Создаем DataFrame из данных о домах
houses_df = pd.DataFrame(houses_data)

# Выводим первые строки DataFrame
print("Информация о домах:")
display(houses_df.head())
```

```python
# URL для запроса домов с девизом
houses_with_words_url = "https://www.anapioficeandfire.com/api/houses?hasWords=True"

# Получаем данные о домах с девизом
response_houses_with_words = requests.get(houses_with_words_url)
if response_houses_with_words.status_code == 200:
    houses_with_words_data = response_houses_with_words.json()  # Преобразуем ответ в JSON
else:
    print(f"Ошибка при запросе домов с девизом: {response_houses_with_words.status_code}")
    houses_with_words_data = []

# Создаем DataFrame из данных о домах с девизом
houses_with_words_df = pd.DataFrame(houses_with_words_data)

# Выводим первые строки DataFrame
print("Информация о домах с девизом:")
display(houses_with_words_df.head())
```

```python
import pandas as pd

# Создаем курсор для выполнения запросов
cursor = conn.cursor()

# Выполняем SQL-запрос
query_all = "SELECT * FROM rnc_database LIMIT 10;"
cursor.execute(query_all)

# Получаем результаты
rows = cursor.fetchall()

# Преобразуем результаты в DataFrame
columns = [desc[0] for desc in cursor.description]  # Имена столбцов
df_all = pd.DataFrame(rows, columns=columns)

# Выводим первые строки DataFrame
print("Первые 10 строк таблицы rnc_database:")
display(df_all.head())
```

```python
# Выполняем SQL-запрос для выбора конкретных столбцов
query_specific = """
SELECT display_name, num_sequences, num_organisms, url
FROM rnc_database
LIMIT 10;
"""
cursor.execute(query_specific)

# Получаем результаты
specific_rows = cursor.fetchall()

# Преобразуем результаты в DataFrame
specific_columns = ["display_name", "num_sequences", "num_organisms", "url"]
df_specific = pd.DataFrame(specific_rows, columns=specific_columns)

# Выводим первые строки DataFrame
print("Значения столбцов display_name, num_sequences, num_organisms, url:")
display(df_specific.head())
```
